```
import numpy as np
import matplotlib.pyplot as plt
```

```
def loadData():
  X = np.array([[1], [2], [3], [4], [5], [6]])
  y = np.array([[1], [4], [9], [16], [25], [36]])
  return X, y
```

```
def costfunc(X, theta, y):
  cost = np.sum(np.subtract(X @ theta,y)**2)/(2*y.size)
  return cost
```

```
def model(X, y, theta, learnRate, epochs):
  costHist = []
  for i in range(epochs):
    grad = (learnRate/y.size)*(X.T @ (X@theta-y))
    theta = np.subtract(theta, grad)
    costHist.append(costfunc(X, theta, y))
  return costHist, theta
```

```
X, y = loadData()
Xnew = np.column_stack((np.ones(y.shape[0]), X, X**2, X**3)) ############ chosing polynomial of degree 3
theta = np.zeros((Xnew.shape[1], 1))
epochs = 1000000
(hist, theta) = model(Xnew, y, theta, 0.00005, epochs)
```
