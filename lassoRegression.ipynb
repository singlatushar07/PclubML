```
import numpy as np
import matplotlib.pyplot as plt
```

```
def loadData():
  X = np.array([[1], [2], [3], [4]])
  y = np.array([[1], [4], [9], [16]])
  return X, y
```

```
def costfunc(X, theta, y, lamda):
  cost = np.sum(np.subtract(X @ theta,y)**2)/(2*y.size)
  theta_abs = np.absolute(theta)
  cost += lamda*theta_abs.sum()
  return cost
```

```
def model(X, y, theta, learnRate, epochs, lamda):
  costHist = []
  for i in range(epochs):
    grad = (learnRate/y.size)*(X.T @ (X@theta-y))
    regularization_term = learnRate*lamda*(np.sign(theta))   ########### minimizing cost
    regularization_term[0][0] = 0               
    theta = np.subtract(theta, grad)
    theta = np.subtract(theta, regularization_term)
    costHist.append(costfunc(X, theta, y, lamda))
  return costHist, theta
```

```
X, y = loadData()
Xnew = np.column_stack((np.ones(y.shape[0]), X, X**2, X**3))  ############# choosing polynomial of degree 3
theta = np.zeros((Xnew.shape[1], 1))
epochs = 10000
(hist, theta) = model(Xnew, y, theta, 0.001, epochs, 100000000)
```
